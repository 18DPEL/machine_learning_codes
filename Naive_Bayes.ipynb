# Meshgrid visualization for 2 features

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_classification
from sklearn.naive_bayes import GaussianNB

# Generate some random data
n_features = 3 # or 4
X, y = make_classification(n_samples=200, n_features=n_features, n_redundant=0, n_informative=n_features,
                           random_state=42, n_clusters_per_class=1, class_sep=1.5)

# Train a Gaussian Naive Bayes classifier
clf = GaussianNB()
clf.fit(X, y)

# Define the range of values for each feature
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
z_min, z_max = X[:, 2].min() - 1, X[:, 2].max() + 1
# Uncomment the following line if you have 4 features
# w_min, w_max = X[:, 3].min() - 1, X[:, 3].max() + 1

# Create a grid of points
xx, yy, zz = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1),
                         np.arange(z_min, z_max, 0.1))
# Uncomment the following line if you have 4 features
# ww, xx, yy, zz = np.meshgrid(np.arange(w_min, w_max, 0.1),
#                               np.arange(x_min, x_max, 0.1),
#                               np.arange(y_min, y_max, 0.1),
#                               np.arange(z_min, z_max, 0.1))

# Predict the class for each point on the grid
Z = clf.predict(np.c_[xx.ravel(), yy.ravel(), zz.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundaries and data points
fig = plt.figure()
if n_features == 3:
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, alpha=0.8)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_zlabel('Feature 3')
    ax.set_title('Naive Bayes Decision Boundaries (3 features)')
    ax.plot_surface(xx, yy, zz, rstride=1, cstride=1, alpha=0.3)
elif n_features == 4:
    ax = fig.add_subplot(111, projection='4d')
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], X[:, 3], c=y, alpha=0.8)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_zlabel('Feature 3')
    ax.set_wlabel('Feature 4')
    ax.set_title('Naive Bayes Decision Boundaries (4 features)')
    ax.plot_surface(ww, xx, yy, zz, rstride=1, cstride=1, alpha=0.3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.naive_bayes import GaussianNB

# Generate some random data
X, y =make_classification(n_samples=300, n_features=2, n_redundant=0, n_informative=2,
                          random_state=42, n_clusters_per_class=1, class_sep=1.5)

# Train a Gaussian Naive Bayes classifier
clf = GaussianNB()
clf.fit(X, y)

# Define the range of values for x and y
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

# Create a grid of points
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Predict the class for each point on the grid
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundaries and data points
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.8)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Naive Bayes Decision Boundaries')
plt.show()
```
